Resumen Ejecutivo: Diagnóstico de Desempeño Comercial (2024–2025)
Empresa: Andes Retail Group

📊 1. Indicadores Clave de Rendimiento (KPIs)
A nivel macro, el desempeño global consolidado durante el periodo de análisis se compone de:
- Ingresos Totales: $6.00M  
- Ganancias Totales: $1.94M (Calculado como Ingresos - Costos)  
- Volumen Operativo: 58K unidades vendidas

📖 3. Narrativa Estratégica (Modelo SCQA)
🖥️ Análisis General (Overview)
- Situación (S): Durante el periodo 2024-2025, el grupo generó $6M en ingresos y $1.94M en ganancias, operando bajo una marcada estacionalidad comercial.  
- Complicación (C): Al comparar los años de forma global, 2025 muestra una contracción en los ingresos. Sin embargo, el análisis mensual revela que de marzo a noviembre el rendimiento es idéntico o superior a 2024; la caída real se concentra exclusivamente en un débil arranque (enero-febrero) y un cierre muy por debajo del periodo anterior.  
- Pregunta (Q): ¿Cómo se distribuye geográficamente la rentabilidad y qué explica esta caída en los meses de apertura y cierre?  
- Respuesta (A): La ganancia territorial está perfectamente equilibrada (liderada sutilmente por la región Norte, seguida de Centro y Sur), lo que demuestra que el problema comercial no es una debilidad constante o geográfica del negocio, sino un problema estrictamente estacional localizado en los extremos del año. 

🔎 Análisis de Causa Raíz (Detalle)
- Situación (S): Se desglosaron las métricas de ingresos por líneas de producto, comportamiento del ticket regional y tendencias temporales por tipo de cliente.  
- Complicación (C): La facturación global demuestra una alta dependencia hacia ciertas categorías y sufre por una falta de tracción crítica en temporadas altas por parte de nichos específicos.  
- Pregunta (Q): ¿Qué categorías sostienen el negocio, si el valor del ticket afecta el estancamiento regional y qué segmento de clientes falló en los meses clave?  
- Respuesta (A):
  - Deportes y Electrónica se consolidan como los pilares fundamentales de facturación.  
  - La proporción del ticket es idéntica en todas las regiones (70% Venta Alta y 30% Venta Baja), descartando anomalías o problemas de poder adquisitivo por territorio.  
  - El Segmento Premium se mantuvo plano y estancado en niveles mínimos todo el año, siendo el responsable directo de que no se impulsaran las ventas en enero, febrero y diciembre, mientras que los segmentos Económico y Estándar sostuvieron toda la estacionalidad.

🎯 4. Conclusiones y Recomendaciones de Negocio
- Diagnóstico: El núcleo del negocio es sano y las regiones responden de manera homogénea en volumen y valor de ticket. La pérdida de ingresos en 2025 se debe al estancamiento del sector Premium durante las temporadas altas de consumo.  
- Plan de Acción:
  - Reestructurar la estrategia de captación y fidelización dirigida específicamente al consumidor del nicho Premium.  
  - Diseñar campañas comerciales anticipadas para blindar los meses críticos de enero, febrero y diciembre, evitando que la fuga estacional se repita en los próximos ejercicios.
